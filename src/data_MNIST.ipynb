### MNIST

```python
import torch
import torchvision
import torchvision.datasets as datasets
from torchvision import transforms
from PIL import Image
```

```python
mnist_trainset = datasets.MNIST(root='../data', train=True, download=True, transform=None)
```

```python
mnist_testset = datasets.MNIST(root='../data', train=False, download=True, transform=None)
```

```python
mnist_trainset[0][0].show() # number 5
```

```python
mnist_trainset[0][1] 
```

```python
X_train = mnist_trainset.data.float()/ 255.0 
X_train = X_train.unsqueeze(1) #(N, 28, 28) -> (N, 1, 28, 28) # We have to add this channel cause most layers in NN expect these 4 dim, so N batch size, C the number of colors(RGB->3 and B/W -> 1 intensité lumineuse), H and W of the images
y_train = mnist_trainset.targets
```

```python
X_test = mnist_testset.data.float()/ 255.0 
X_test= X_test.unsqueeze(1) 
y_test = mnist_testset.targets
```

I will save the processed data for ease of use later

```python
import torch
import os
fold = '../data/processed_data'

if not os.path.exists(fold):
    os.makedirs(fold)
   
data = {
    'X_train': X_train,
    'X_test': X_test,
    'y_train': y_train,
    'y_test': y_test
    }

file_path = os.path.join(fold, 'mnist_processed.pt')
torch.save(data, file_path)
```

This how to get the data from the file

```python
path = os.path.join('../data/processed_data', 'mnist_processed.pt')
data = torch.load(path)
X_train = data['X_train']
X_test = data['X_test']
y_train = data['y_train']
y_test = data['y_test']

```
