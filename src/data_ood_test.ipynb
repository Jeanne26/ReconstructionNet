```python
tissuemnist_trainset = TissueMNIST(split='train', download=True, root='../data', transform=None)
tissuemnist_testset = TissueMNIST(split='test', download=True, root='../data', transform=None)
tissuemnist_valset = TissueMNIST(split='val', download=True, root='../data', transform=None)
```

```python
tissue_X_train = torch.from_numpy(tissuemnist_trainset.imgs).unsqueeze(1).float() / 255.0
tissue_y_train = torch.from_numpy(tissuemnist_trainset.labels).long().squeeze()

tissue_X_test = torch.from_numpy(tissuemnist_testset.imgs).unsqueeze(1).float() / 255.0
tissue_y_test = torch.from_numpy(tissuemnist_testset.labels).long().squeeze()

tissue_X_val = torch.from_numpy(tissuemnist_valset.imgs).unsqueeze(1).float() / 255.0
tissue_y_val = torch.from_numpy(tissuemnist_valset.labels).long().squeeze()
```

```python
# Save TissueMNIST
data = {
    'X_train': tissue_X_train,
    'X_test': tissue_X_test,
    'y_train': tissue_y_train,
    'y_test': tissue_y_test,
    'X_val': tissue_X_val,
    'y_val': tissue_y_val
}

file_path = os.path.join(fold, 'tissuemnist_processed.pt')
torch.save(data, file_path)
print(f"Saved to {file_path}")
```

```python
# Load ChestMNIST
chestmnist_trainset = ChestMNIST(split='train', download=True, root='../data', transform=None)
chestmnist_testset = ChestMNIST(split='test', download=True, root='../data', transform=None)
chestmnist_valset = ChestMNIST(split='val', download=True, root='../data', transform=None)
```

```python
# Convert and normalize ChestMNIST
chest_X_train = torch.from_numpy(chestmnist_trainset.imgs).unsqueeze(1).float() / 255.0
chest_y_train = torch.from_numpy(chestmnist_trainset.labels).long().squeeze()

chest_X_test = torch.from_numpy(chestmnist_testset.imgs).unsqueeze(1).float() / 255.0
chest_y_test = torch.from_numpy(chestmnist_testset.labels).long().squeeze()

chest_X_val = torch.from_numpy(chestmnist_valset.imgs).unsqueeze(1).float() / 255.0
chest_y_val = torch.from_numpy(chestmnist_valset.labels).long().squeeze()

```

```python
# Save ChestMNIST
fold = '../data/processed_data'
if not os.path.exists(fold):
    os.makedirs(fold)

data = {
    'X_train': chest_X_train,
    'X_test': chest_X_test,
    'y_train': chest_y_train,
    'y_test': chest_y_test,
    'X_val': chest_X_val,
    'y_val': chest_y_val
}

file_path = os.path.join(fold, 'chestmnist_processed.pt')
torch.save(data, file_path)
print(f"Saved to {file_path}")
```
