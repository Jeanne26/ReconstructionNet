```python
from torchvision import transforms
from PIL import Image
from medmnist import OCTMNIST

data_transform = transforms.Compose([
    transforms.ToTensor(),  # Converts to tensor and scales to [0, 1] instead of what we did with MNIST here I figured out I can do that directly (also this means that I dont need to use unsqueze to add the channels dim)
])
```

```python
octmnist_trainset = OCTMNIST(split='train', download=True, root='../data', transform=data_transform)
```

```python
octmnist_testset = OCTMNIST(split='test', download=True, root='../data', transform=data_transform)
```

```python
octmnist_valset = OCTMNIST(split='val', download=True, root='../data', transform=data_transform)
```

```python
X_train = octmnist_trainset.imgs
y_train = octmnist_trainset.labels

X_test = octmnist_testset.imgs
y_test = octmnist_testset.labels

X_val = octmnist_valset.imgs
y_val = octmnist_valset.labels

```

```python
import torch
import os
fold = '../data/processed_data'

if not os.path.exists(fold):
    os.makedirs(fold)
   
data = {
    'X_train': X_train,
    'X_test': X_test,
    'y_train': y_train,
    'y_test': y_test,
    'X_val': X_val,
    'y_val': y_val

    }

file_path = os.path.join(fold, 'octmnist_processed.pt')
torch.save(data, file_path)
```

To get the data :

```python
path = os.path.join('../data/processed_data', 'octmnist_processed.pt')
data = torch.load(path, weights_only=False)
X_train = data['X_train']
X_test = data['X_test']
y_train = data['y_train']
y_test = data['y_test']
X_val = data['X_val']
y_val = data['y_val']
```
